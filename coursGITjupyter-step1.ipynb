```python
import serial
import time

PORT = "/dev/tty.usbmodem34B7DA607EBC2"  # adapte selon ton PC
BAUDRATE = 1000000

print("Connexion au port série...")
ser = serial.Serial(PORT, BAUDRATE, timeout=1)
time.sleep(2)  # attendre que l'Arduino soit prêt

# 1. Démarrage
print("Envoi du caractère START (#)...")
ser.write(b"#")

start_time = time.time()

try:
    while True:
        line = ser.readline().decode(errors='ignore').strip()
        if line:
            print(f"> {line}")

        # 2. Après 10 secondes, demander un reset
        if time.time() - start_time > 10:
            print("Envoi du caractère RESET (!)...")
            ser.write(b"!")
            start_time = float('inf')  # éviter de renvoyer plusieurs fois

except KeyboardInterrupt:
    print("\nArrêt manuel.")

finally:
    ser.close()
```
